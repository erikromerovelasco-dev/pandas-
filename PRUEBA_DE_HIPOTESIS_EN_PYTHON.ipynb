```python
# problema 1 tornillos

import numpy as np
import scipy.stats as stats
import pandas as pd

mu_0 = 10
n = 49
x_bar = 9.7
s = 0.5
alpha = 0.01

z = (x_bar - mu_0) / (s / np.sqrt(n))

z_critical = stats.norm.ppf(alpha)

if z < z_critical:
    conclusion = "Rechazar la hipótesis nula"
else:
    conclusion = "No rechazar la hipótesis nula"

df = pd.DataFrame({
    'parametros': ['Media teorica','Media muestral','Desviación estándar','Tamaño de la muestra','Z-Calculado','Z-Crítico','Conclusión'],
    'valores': [mu_0,x_bar,s,n,z,z_critical,conclusion]
})

print(df)
```
