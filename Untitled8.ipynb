```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
from google.colab import files
# uploaded = files.upload() # No es necesario si el archivo ya está en el entorno
```

```python
df = pd.read_csv('PracticaExamenes.csv')

print(df.head())
```

```python
X = df[['horas']]
y = df['puntuacion']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
m = modelo.coef_[0]
b = modelo.intercept_

print(f"Ecuación: y = {m:.2f}x + {b:.2f}")
```

```python
y_pred = modelo.predict(X)
r2 = r2_score(y, y_pred)
print(f"R²: {r2}")
```

```python
if r2 > 0.7:
    print("Alta relación entre horas y puntaje")
elif r2 > 0.4:
    print("Relación moderada")
else:
    print("Relación débil")
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("Horas de práctica")
plt.ylabel("Puntaje examen")
plt.title("Regresión: Horas vs Puntaje")
plt.show()
```

```python
def predecir_puntaje(horas):
    return modelo.predict([[horas]])[0]

print("Ejemplo:", predecir_puntaje(10))
```
