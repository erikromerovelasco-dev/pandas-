```python
import numpy as np
from scipy import stats

# Ejercicio 1
tabla1 = np.array([[85, 65],
                   [50, 100]])

chi2, p_value, gl, esperados = stats.chi2_contingency(tabla1, correction=False)

print("Ejercicio 1")
print(f"Chi-cuadrado = {chi2:.4f}")
print(f"p-value = {p_value:.4f}")
print(f"Grados de libertad = {gl}")
print("Frecuencias esperadas:")
print(esperados)

# Assuming 'alpha' is defined elsewhere or should be here
alpha = 0.05 # Add alpha definition if it's not guaranteed to be run from another cell

if p_value < alpha:
    print("Conclusión: Hay relación entre la ciudad y el transporte.\n")
else:
    print("Conclusión: No hay relación entre la ciudad y el transporte.\n")
```
