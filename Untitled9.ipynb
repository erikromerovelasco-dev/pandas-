```python
#ejercicio 3
import numpy as np
from scipy import stats

antes = np.array([220, 210, 230, 240, 215, 225, 235, 200, 210, 220,
                  230, 215, 225, 240, 235, 210, 220, 230, 225, 215])

despues = np.array([200, 195, 210, 220, 200, 210, 215, 190, 200, 205,
                    210, 200, 210, 220, 215, 195, 205, 210, 205, 200])

t_stat, p_value = stats.ttest_rel(antes, despues)

print("t =", t_stat)
print("p =", p_value)


alpha = 0.05
if p_value < alpha:
    print("Se rechaza H0: Sí hubo reducción significativa")
else:
    print("No se rechaza H0: No hay evidencia suficiente")
```

```python
#ejercicio 4
import numpy as np
from scipy import stats

# Datos simulados
jovenes = np.array([250, 260, 245, 255, 240, 235, 260, 250, 245, 255,
                    240, 250, 260, 245, 255, 240, 235, 250, 245, 255])

adultos = np.array([300, 310, 320, 305, 295, 315, 300, 310, 305, 295,
                    320, 310, 305, 300, 315, 320, 310, 305, 300, 315])

# Prueba t independiente
t_stat, p_value = stats.ttest_ind(jovenes, adultos)

print("t =", t_stat)
print("p =", p_value)

# Decisión
alpha = 0.05
if p_value < alpha:
    print("Se rechaza H0: Sí hay diferencia significativa")
else:
    print("No se rechaza H0: No hay diferencia significativa")
```
