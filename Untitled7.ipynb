```python
import pandas as pd
df = pd.read_csv('EdadSeguros.csv')

print(df.head())
```

```python
X = df[['Edad']]
y = df['Costo']
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X, y)
plt.xlabel("Edad")
plt.ylabel("Costo mensual")
plt.title("Relación Edad vs Costo")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
print("Pendiente:", modelo.coef_[0])
print("Intercepto:", modelo.intercept_)
```

```python
import statsmodels.api as sm
X_sm = sm.add_constant(X)
modelo_sm = sm.OLS(y, X_sm).fit()

print(modelo_sm.summary())
```

```python
pred = modelo_sm.get_prediction(X_sm)
pred_summary = pred.summary_frame(alpha=0.05)
```

```python
plt.figure()
plt.scatter(X, y)
plt.plot(X, modelo.predict(X))
plt.xlabel("Edad")
plt.ylabel("Costo mensual")
plt.title("Regresión Edad vs Costo")
plt.show()
```

```python
edades = pd.DataFrame({'Edad': [30, 40, 50, 60]})
predicciones = modelo.predict(edades)

tabla = edades.copy()
tabla['Costo estimado'] = predicciones

print(tabla)
```
