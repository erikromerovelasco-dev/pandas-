```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('PublicidadVentas.csv')
```

```python
print(df.head())
```

```python
X = df[['Inversion']]
y = df['Ventas']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
y_pred = modelo.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
print(f"MSE: {mse}")
print(f"R²: {r2}")
```

```python
def predecir_ventas(inversion):
    return modelo.predict([[inversion]])[0]

print("Ejemplo predicción:", predecir_ventas(50))

# Gráfico
plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("Inversión en publicidad (miles de $)")
plt.ylabel("Ventas")
plt.title("Regresión lineal: Publicidad vs Ventas")
```

```python
m = modelo.coef_[0]
b = modelo.intercept_
plt.text(min(X.values), max(y), f'y = {m:.2f}x + {b:.2f}')

plt.show()
```
