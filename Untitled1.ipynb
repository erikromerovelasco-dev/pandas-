```python
import pandas as pd
df=pd.read_csv("part1_employees.csv")
df.head()
```

```python
x=(df["perf_score"]>=0.85)&(df["dept"]=="Eng")
df.loc[x,["emp_id","name","dept","city","perf_score","salary_mxn"]]
```

```python
result=df.query("(city == 'CDMX' or city == 'GDL') and warnings == 0 and absences <= 1")[["emp_id","name","city","warnings","absences"]]
result
```

```python
df.agg({
    "salary_mxn": "mean",
    "perf_score": "mean",
    "warnings": "sum"
})
```

```python
dept=df.groupby('dept').agg(
    num_empleados=('emp_id', 'size'),
    salario_promedio=('salary_mxn', 'mean'),
    perf_promedio=('perf_score', 'mean')
)

dept_sorted = dept.sort_values(by='salario_promedio', ascending=False)
dept_sorted.head()
```
