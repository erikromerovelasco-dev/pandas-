```python
# ==========================================
# EJERCICIO 1 - KMEANS CLIENTES
# ==========================================

# Librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# -------------------------
# 1. Crear DataFrame
# -------------------------

compras = [1, 2, 1.5, 10, 12, 11, 4, 5, 6]
gasto = [2, 3, 2.5, 9, 11, 10, 5, 6, 7]

df = pd.DataFrame({
    'Compras': compras,
    'Gasto': gasto
})

print("DataFrame:")
print(df)

# -------------------------
# 2. Graficar puntos
# -------------------------

plt.figure(figsize=(6,5))
plt.scatter(df['Compras'], df['Gasto'])

plt.title('Clientes')
plt.xlabel('Compras al mes')
plt.ylabel('Gasto mensual')

plt.grid()
plt.show()

# -------------------------
# 3. Método del codo
# -------------------------

inercias = []

for k in range(1, 6):
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(df)

    inercias.append(modelo.inertia_)

plt.figure(figsize=(6,5))
plt.plot(range(1,6), inercias, marker='o')

plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')

plt.grid()
plt.show()

# -------------------------
# Aplicar K-Means
# -------------------------

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# -------------------------
# 4. Mostrar clusters
# -------------------------

print("\nDataFrame con Clusters:")
print(df)

# -------------------------
# 5. Mostrar centroides
# -------------------------

centroides = kmeans.cluster_centers_

print("\nCentroides:")
print(centroides)

# -------------------------
# 6. Graficar clusters
# -------------------------

plt.figure(figsize=(7,6))

plt.scatter(df['Compras'],
            df['Gasto'],
            c=df['Cluster'],
            cmap='viridis',
            s=100)

plt.scatter(centroides[:,0],
            centroides[:,1],
            color='red',
            marker='X',
            s=300,
            label='Centroides')

plt.title('Clusters de Clientes')
plt.xlabel('Compras al mes')
plt.ylabel('Gasto mensual')

plt.legend()
plt.grid()
plt.show()
```

```python
# ==========================================
# EJERCICIO 2 - KMEANS SALUD FÍSICA
# ==========================================

# Librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# -------------------------
# 1. Crear DataFrame
# -------------------------

ejercicio = [0.5, 1, 1.5, 6, 7, 8, 2, 3, 4]
condicion = [2, 3, 2.5, 8, 9, 10, 4, 5, 6]

df = pd.DataFrame({
    'Ejercicio': ejercicio,
    'Condicion': condicion
})

print("DataFrame:")
print(df)

# -------------------------
# 2. Visualizar datos
# -------------------------

plt.figure(figsize=(6,5))
plt.scatter(df['Ejercicio'], df['Condicion'])

plt.title('Condición Física')
plt.xlabel('Horas de ejercicio')
plt.ylabel('Nivel de condición')

plt.grid()
plt.show()

# -------------------------
# 3. Método del codo
# -------------------------

inercias = []

for k in range(1,6):
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(df)

    inercias.append(modelo.inertia_)

plt.figure(figsize=(6,5))
plt.plot(range(1,6), inercias, marker='o')

plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')

plt.grid()
plt.show()

# -------------------------
# Aplicar K-Means
# -------------------------

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# -------------------------
# 4. Mostrar clusters
# -------------------------

print("\nClusters:")
print(df)

# -------------------------
# 5. Obtener centroides
# -------------------------

centroides = kmeans.cluster_centers_

print("\nCentroides:")
print(centroides)

# -------------------------
# 6. Graficar resultados
# -------------------------

plt.figure(figsize=(7,6))

plt.scatter(df['Ejercicio'],
            df['Condicion'],
            c=df['Cluster'],
            cmap='plasma',
            s=100)

plt.scatter(centroides[:,0],
            centroides[:,1],
            color='red',
            marker='X',
            s=300,
            label='Centroides')

plt.title('Clusters de Condición Física')
plt.xlabel('Horas de ejercicio')
plt.ylabel('Condición física')

plt.legend()
plt.grid()
plt.show()
```

```python
# ==========================================
# EJERCICIO 3 - KMEANS PRODUCTOS
# ==========================================

# Librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# -------------------------
# 1. Crear DataFrame
# -------------------------

ventas = [2, 3, 2.5, 15, 18, 20, 5, 6, 7]
ingresos = [3, 4, 3.5, 9, 10, 10, 5, 6, 7]

df = pd.DataFrame({
    'Ventas': ventas,
    'Ingresos': ingresos
})

print("DataFrame:")
print(df)

# -------------------------
# 2. Graficar datos
# -------------------------

plt.figure(figsize=(6,5))
plt.scatter(df['Ventas'], df['Ingresos'])

plt.title('Productos')
plt.xlabel('Ventas mensuales')
plt.ylabel('Ingresos')

plt.grid()
plt.show()

# -------------------------
# 3. Aplicar K-Means
# -------------------------

kmeans = KMeans(n_clusters=3, random_state=42)

df['Cluster'] = kmeans.fit_predict(df)

# -------------------------
# 4. Mostrar clusters
# -------------------------

print("\nClusters:")
print(df)

# -------------------------
# 5. Obtener centroides
# -------------------------

centroides = kmeans.cluster_centers_

print("\nCentroides:")
print(centroides)

# -------------------------
# 6. Visualizar clusters
# -------------------------

plt.figure(figsize=(7,6))

plt.scatter(df['Ventas'],
            df['Ingresos'],
            c=df['Cluster'],
            cmap='rainbow',
            s=100)

plt.scatter(centroides[:,0],
            centroides[:,1],
            color='black',
            marker='X',
            s=300,
            label='Centroides')

plt.title('Clusters de Productos')
plt.xlabel('Ventas mensuales')
plt.ylabel('Ingresos')

plt.legend()
plt.grid()
plt.show()
```
